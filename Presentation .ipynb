```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
from IPython.html.widgets import interact, fixed
import Equation_Sheet as es
gamma=4.5e-8
```

#The equations used to track the stars and galaxies:

$$ \ddot{\mathbf{r}} = -\gamma \left\{ \frac{M}{r^3}\mathbf{r} -\frac{S}{\rho^3}\boldsymbol{\rho} + \frac{S}{R^3}\boldsymbol\Re \right\} $$

$$ \ddot{\boldsymbol\Re} = -\gamma \frac{M+S}{R^3}\boldsymbol\Re$$

#Units:

Time: Ages ($10^{8}$years)

Length: Kiloparsecs

Mass: Solar Mass Units

#Base Code

```python
def derivs2(ivec,t,M,S):
        x=ivec[0]
        vx=ivec[1]
        y=ivec[2]
        vy=ivec[3]
                        #this calls the parts of the "ivec" which is my solution vector
        X=ivec[4]
        Vx=ivec[5]
        Y=ivec[6]
        Vy=ivec[7]
        
        dx=vx
        dy=vy
                        #redefines the velocity
        dX=Vx
        dY=Vy
        
        r=np.sqrt(x**2+y**2)
        R=np.sqrt(X**2+Y**2)      
                                        #finds the magnitude of the different inputs
        rho=np.sqrt((X-x)**2+(Y-y)**2)
        rhoy=Y-y
        rhox=X-x
        
        dvx = -gamma*((M/r**3)*x-(S/rho**3)*rhox+(S/R**3)*X)
        dVx = -gamma*((M+S)/R**3)*X
                                            #partial differential equations
        dvy = -gamma*((M/r**3)*y-(S/rho**3)*rhoy+(S/R**3)*Y)
        dVy = -gamma*((M+S)/R**3)*Y
        return np.array([dx,dvx,dy,dvy,dX,dVx,dY,dVy])
```

```python
def sol(ic, max_t, ntimes, M, S):
    t = np.linspace(0.0, max_t, ntimes)
    s = odeint(derivs2, ic, t, args=(M,S)) #solves the differential equation
    return s
```

```python
def generate_star_ics(M, radii, nstars):
    ics = []
    for i in range(len(radii)):
        r = radii[i]
        N = nstars[i]
        theta = np.arange(0.0, 2.0*np.pi, 2.0*np.pi/N)
        for t in range(N):
            x = r*np.cos(theta[t])  #these are the x and y positions
            y = r*np.sin(theta[t])  
            v = np.sqrt((gamma*M)/r) #the magnitude of the velocity vector
            vx = -v*np.sin(theta[t]) #the velocities in the x andy directions
            vy = v*np.cos(theta[t])
            ics.append(np.array([x,vx,y,vy]))
    return ics
```

```python
def generate_gala_ics(M, S, r0, m):
    Y = m*r0                                    #generates the x and y positions of the converging galaxy
    X = (-Y**2.0)/(4.0*r0)+r0
    V=np.sqrt(2.0*gamma*(M+S)/np.sqrt(X**2+Y**2))
    Vx = V*np.cos(angle(Y,r0))                  #generates the x and y velocities of the converging galaxy
    Vy = -V*np.sin(angle(Y,r0))
    return np.array([X,Vx,Y,Vy])
```

```python
def make_stars(M, S, radii, nstars, r0, m, max_t, ntimes):
    star_ics = generate_star_ics(M,radii,nstars)
    gala_ics = generate_gala_ics(M,S,r0,m)
    starx = []
    stary = []
    for sic in star_ics:
        ic = np.hstack([sic, gala_ics])        #takes each star and stiches it to the galaxy
        result = sol( ic, max_t, ntimes, M, S)
        starx.append(result[:,0])
        stary.append(result[:,2])
        galax = result[:,4]
        galay = result[:,6]
    starx = np.transpose(np.array(starx))
    stary = np.transpose(np.array(stary))
    return starx, stary, galax, galay
```

```python
def plot_solution(starx, stary, galax, galay, j, lim):
    px=np.linspace(-100,100,100)   #plots the path of the converging galaxy
    r0=25.0
    py=-px**2/(4.0*r0)+r0
    plt.plot(py,px,color='orchid')
    plt.scatter(starx[j],stary[j],color='b')
    plt.scatter(galax[j],galay[j],color='lime')
    plt.scatter(0,0,color='r')
    plt.xlim(-lim,lim)
    plt.ylim(-lim,lim)
```
